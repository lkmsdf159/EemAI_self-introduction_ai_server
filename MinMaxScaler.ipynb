```python
data = pd.read_csv("/content/BreastCancer_Prediction.csv")
data.head(10)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

scaler = MinMaxScaler()
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])

print("Min-Max Normalized Data:")
print(data.head())
```

```python
hidden_layer1_neurons = 30
hidden_layer2_neurons = 50
output_neurons = 1
activation_hidden_layers = 'relu'
activation_output_layer = 'sigmoid'
optimizer = Adam()
loss_function = 'binary_crossentropy'
performance_metric = 'accuracy'
random_state = 5
epochs = 10
```

```python
X = np.random.rand(1000, 10)
y = np.random.randint(0, 2, size=(1000, 1))
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=random_state)
```

```python
model = Sequential()
model.add(Dense(hidden_layer1_neurons, input_shape=(X.shape[1],), activation=activation_hidden_layers))
model.add(Dense(hidden_layer2_neurons, activation=activation_hidden_layers))
model.add(Dense(output_neurons, activation=activation_output_layer))
```

```python
model.compile(optimizer=optimizer, loss=loss_function, metrics=[performance_metric])
model.summary()
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=epochs, verbose=1)
```

```python
test_loss, test_accuracy = model.evaluate(X_val, y_val)
print(f"loss: {test_loss}")
print(f"accuracy: {test_accuracy}")
```
