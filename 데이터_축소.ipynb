```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.decomposition import PCA

data = pd.read_csv('/content/diabetes (2).csv')
data.sample(10)
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
Y=data['Outcome']
X=data
X.drop(['Outcome'], axis = 1, inplace = True)
X.sample(10)
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
std = std.fit_transform(X)
X
```

```python
pca_diabetes = PCA(n_components = 2)
PrinciplComp_diabetes = pca_diabetes.fit_transform(X)
PrinciplComp_diabetes
```

```python
PrinciplComp_diabetes = pd.DataFrame(data = PrinciplComp_diabetes, columns = ['principal component 1', 'principal component 2'])
PrinciplComp_diabetes.sample(10)
```

```python
print('Explained variation per principal component: {}'.format(pca_diabetes.explained_variance_ratio_))
```
